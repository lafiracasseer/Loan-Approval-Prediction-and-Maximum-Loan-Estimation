---



## Code Reuse Session 1 - Prompt 16

```python
#box plot for the age variable to visually detect outliers using IQR
age_box_fig = px.box(data, x='age')
# Display the box plot
age_box_fig.show()
```

## Code Reuse Session 1 - Prompt 17

```python
#Building IQR outlier detection function

def find_outliers_IQR(data):

  #first quartile(25th percentile)
  q1=data.quantile(0.25)

  #third quartile(75th percentile)
  q3=data.quantile(0.75)

  #interquartile range
  IQR=q3-q1

  #outliers below Q1-1.5*IQR or above Q3+1.5*IQR
  outliers = data[((data<(q1-1.5*IQR))|(data > (q3+1.5*IQR)))]
  return outliers
```

## Code Reuse Session 1 - Prompt 19

```python
#Detecting outliers based on IQR

#Identifying outliers in age
outliers = find_outliers_IQR(data['age'])
print("number of outliers: "+ str(len(outliers)))

#Display the outlier values
outliers
```

```python
#Caling  the IQR function to find outliers in the max_allowed_loan variable
outliers_loan = find_outliers_IQR(data['max_allowed_loan'])
print("number of outliers: " + str(len(outliers_loan)))

#Displaying the outlier values
outliers_loan
```

# **CASE STUDY B : Regression Dataset Preparation**

Creating a regression dataset for filter approved clients


---



## Code Reuse Session 1 - Prompt 28

```python
#reloading the original dataset to retain max_allowed_loan for regression
data_reg = pd.read_csv('/content/drive/MyDrive/loan_approval_data.csv')

#Dropping the id column (non-predictive identifier)
data_reg = data_reg.drop(columns=(['id']))

#Removeing the extreme age outlier (age=123)
data_reg.drop(data_reg.index[[47336]], inplace=True)

#Performing mean imputation for age
Mean_age_reg = data_reg['age'].mean()
data_reg['age'].fillna(Mean_age_reg, inplace=True)

#Performing mean imputation for loan_interest_rate
Mean_rate_reg = data_reg['loan_interest_rate'].mean()
data_reg['loan_interest_rate'].fillna(Mean_rate_reg, inplace=True)

#performing casewise deletion for remaining missing values
data_reg = data_reg.dropna()

#Removeing the 3 rows with negative max_allowed_loan (data entry errors)
data_reg.drop(data_reg.index[[0, 1, 2]], inplace=True)

#Filtering to keep only Approved clients (loan_approval_status == 1) for regression
data_reg = data_reg[data_reg['loan_approval_status']<1]

#Dropping loan_approval_status (it is the classification target, not a regression feature)
data_reg = data_reg.drop(columns=['loan_approval_status'])

#Confirming the regression dataset dimensions
print('Regression dataset shape:', data_reg.shape)
print('Features:', list(data_reg.columns))
```

## Code Reuse Session 1 - Prompt 23 and 26

```python
#Applying encoding to regression dataset

#Convertung float columns to integer in regression dataset
data_reg = data_reg.astype({'age': 'int', 'loan_interest_rate': 'int'})

#Mapping payment_default_on_file binary variable to numeric (Y=1, N=0)
data_reg['payment_default_on_file'] = data_reg['payment_default_on_file'].map({'Y': 1, 'N': 0})

#Encoding home_ownership with LabelEncoder
label_encoder = preprocessing.LabelEncoder()
data_reg['home_ownership'] = label_encoder.fit_transform(data_reg['home_ownership'])

#encoding loan_intent with LabelEncoder
label_encoder = preprocessing.LabelEncoder()
data_reg['loan_intent'] = label_encoder.fit_transform(data_reg['loan_intent'])

#Confirming if all columns are numeric
data_reg.dtypes
```

## Code Reuse Session 1 - Prompt 32 and 33

```python
#Importing MinMaxScaler from sklearn preprocessing
from sklearn.preprocessing import MinMaxScaler

#Defining numeric columns for the regression dataset (excluding the target max_allowed_loan)
numeric_cols_reg = ['age', 'income', 'employment_length', 'loan_amount',
                    'loan_interest_rate', 'loan_income_ratio', 'credit_history_length']

#Instantiate the MinMaxScaler object
scaler_reg = MinMaxScaler()

#Applying Min-Max scaling to all numeric columns in the regression dataset
data_reg[numeric_cols_reg] = scaler_reg.fit_transform(data_reg[numeric_cols_reg])

#Verifying scaling – all values should be between 0 and 1
data_reg[numeric_cols_reg].describe()
```
